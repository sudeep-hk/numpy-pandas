```python
vk=delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
def run_score(batsman_name):
    vk=delivery[delivery['batsman']==batsman_name]
    return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(5)
```

```python
run_score('V Kohli')
```

```python
run_score('MS Dhoni')
```
